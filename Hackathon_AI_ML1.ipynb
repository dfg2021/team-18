```python
##using mean to imput the missing values 
new_data =data[['TimePeriod', 'Value', 'Units']].fillna(data['Value'].median())
new_data.head()
```

```python
##Training set up to 2018
train, test= new_data[(new_data['TimePeriod']<=2018) & (new_data['Units']=='PERCENT')], new_data[(new_data['TimePeriod']==2019) & (new_data['Units']=='PERCENT')]
X_train, X_test =np.arange(0, len(train), 1).reshape((-1,1)), np.arange(0, len(test), 1).reshape((-1,1))
y_train, y_test =np.array(train)[:,1:2], np.array(test)[:,1:2]
y_train
```

#### Linear Regression Model

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler =MinMaxScaler()
```

```python
y_train_scaled =scaler.fit_transform(y_train)
```

```python
regressor =LinearRegression()
```

```python
regressor.fit(X_train, y_train_scaled)
```

```python
y_pred =regressor.predict(X_test)
```

```python
y_pred_scaled =scaler.inverse_transform(y_pred)
```

```python
plt.plot(y_pred_scaled, 'r')
plt.scatter(X_test, y_test)
plt.savefig('lin.jpg')
plt.show()
```

### Evaluate the Model

```python
MAE = mean_absolute_error(y_test,y_pred_scaled)
MSE = mean_squared_error(y_test,y_pred_scaled)
RMSE = np.sqrt(MSE)
```

```python
print ('MAE', MAE)
print('MSE', MSE)
print('RMSE', RMSE)
```

### ARIMA

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
#TSA from statsmodels
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels.tsa.api as smt
#Display and Plotting
import seaborn as sns
```

```python
def arim(): 
    train =y_train
    history = [x for x in train]
    test = y_test
    predictions = []
    for t in range(len(test)):
        model = sm.tsa.ARIMA(history, order=(5,1,0))
        model_fit = model.fit(disp=0)
        output = model_fit.forecast()
        yhat = output[0]
        predictions.append(yhat)
        obs = test[t]
        history.append(obs)
        print('predicted=%f, expected=%f' % (yhat, obs))
    return test,predictions
```

```python
yt, yp=arim()
plt.plot(yt, 'ob')
plt.plot(yp, 'or')
plt.xlabel('TimePeriod (2019)')
plt.ylabel('Value')
plt.savefig('arrima.jpg')
plt.show()
```
